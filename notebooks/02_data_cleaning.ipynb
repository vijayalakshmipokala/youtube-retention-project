##Missing Values Handling
- description column had ~570 missing values
- These rows were removed since text is required for sentiment analysis

```python
#handling missing values
df = df.dropna(subset=['description'])
```

```python
# remove duplicates
df = df.drop_duplicates()
```

```python
# remove invalid videos
df = df[df['video_error_or_removed'] == False]
```

```python
#convert datatypes

df['publish_time'] = pd.to_datetime(df['publish_time'], errors='coerce')
df = df.dropna(subset=['publish_time'])
```

## Cleaning Steps Performed
- Removed duplicate records
- Dropped rows with missing descriptions
- Filtered out invalid/removed videos
- Converted `publish_time` to datetime format

```python
# handling outliers

upper_limit = df['views'].quantile(0.99)
df = df[df['views'] < upper_limit]
```

## Outlier Handling
- Top 1% of videos by views were removed
- This prevents extreme viral videos from biasing the analysis
- Ensures more realistic performance distribution
