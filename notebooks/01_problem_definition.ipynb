Problem Statement:
Users often leave the platform when irrelevant or low-engagement videos are recommended. This reduces overall watch time and user retention. The goal is to design a data-driven system that identifies and recommends videos most likely to keep users engaged.

Objective:
Build a recommendation system that ranks videos based on engagement, popularity, sentiment, and timing signals to maximize user retention and watch time.

Key Metrics:

1.Engagement Rate = (likes + comment_count) / views

2.Like Ratio = likes / views

3.Comment Ratio = comment_count / views

4.Views = proxy for popularity and watch time

5.Sentiment Score = audience reaction quality (positive vs negative)

6.Final Score = weighted combination of all signals for ranking videos

Business Questions:

1.Which content categories drive the highest engagement?

2.What publishing time leads to maximum user interaction?

3.Do high-view videos always result in high retention?

4.How does sentiment impact video performance?

5.Can combining multiple signals improve recommendation quality over views alone?


Assumptions:

1.Views are used as a proxy for watch time

2.Likes and comments reflect user engagement

3.Sentiment score reflects user satisfaction

4.Trending dataset represents real-world user behavior patterns


Expected Output:
A ranked list of videos optimized for user retention, along with insights on what drives engagement and performance.

Project Outcome:
This project aims to simulate how a platform like YouTube prioritizes content using data-driven signals, helping improve recommendation quality and maximize user retention.
