# Exploratory Data Analysis - YouTube Retention Project

## Objective
Analyze patterns in video performance to understand what drives engagement and retention.

```python
import sys
sys.path.append("../src")

import pandas as pd
import matplotlib.pyplot as plt

from feature_engineering import feature_engineering_pipeline
from scoring_model import scoring_pipeline

df = pd.read_csv("../data/processed/cleaned_videos.csv")

df = feature_engineering_pipeline(df)
df = scoring_pipeline(df)
```

```python
# Category VS Engagement
category_engagement = df.groupby('category_id')['engagement_rate'].mean().sort_values(ascending=False)

category_engagement.plot(kind='bar')
plt.title("Category vs Engagement Rate")
plt.show()
```

## Insight: Category Impact
Certain categories consistently achieve higher engagement rates, indicating stronger audience interest and retention potential.

```python
# Hours VS Engagement
hour_engagement = df.groupby('hour')['engagement_rate'].mean()

hour_engagement.plot()
plt.title("Hour vs Engagement Rate")
plt.show()
```

## Insight: Time Impact
Videos published during evening hours show higher engagement, suggesting optimal posting windows.

```python
# Sentiment VS Score

plt.scatter(df['sentiment_score'], df['score'])
plt.xlabel("Sentiment Score")
plt.ylabel("Final Score")
plt.title("Sentiment vs Score")
plt.show()
```

## Insight: Sentiment Effect
Positive sentiment is generally associated with higher scores, indicating better audience reception.

```python
# Views VS Engagement

plt.scatter(df['views'], df['engagement_rate'])
plt.xlabel("Views")
plt.ylabel("Engagement Rate")
plt.title("Views vs Engagement")
plt.show()
```

## Insight: Views vs Engagement
High views do not always guarantee high engagement, reinforcing the need for a multi-factor scoring model.
